```python
# Update 'date' to datetime format
df_gundeaths['date'] = pd.to_datetime(df_gundeaths['date'])
```

```python
# Drop 'address' column as it's not relevant to analysis.
if 'address' in df_gundeaths.columns:
    df_gundeaths = df_gundeaths.drop(columns=['address'])

if 'congressional_district' in df_gundeaths.columns:
    df_gundeaths = df_gundeaths.drop(columns=['congressional_district'])

if 'longitude' in df_gundeaths.columns:
    df_gundeaths = df_gundeaths.drop(columns=['longitude'])

if 'participant_name' in df_gundeaths.columns:
    df_gundeaths = df_gundeaths.drop(columns=['participant_name'])

if 'participant_relationship' in df_gundeaths.columns:
    df_gundeaths = df_gundeaths.drop(columns=['participant_relationship'])


if 'participant_status' in df_gundeaths.columns:
    df_gundeaths = df_gundeaths.drop(columns=['participant_status'])

if 'participant_type' in df_gundeaths.columns:
    df_gundeaths = df_gundeaths.drop(columns=['participant_type'])

if 'state_house_district' in df_gundeaths.columns:
    df_gundeaths = df_gundeaths.drop(columns=['state_house_district'])

if 'state_senate_district' in df_gundeaths.columns:
    df_gundeaths = df_gundeaths.drop(columns=['state_senate_district'])

if 'gun_stolen' in df_gundeaths.columns:
    df_gundeaths = df_gundeaths.drop(columns=['gun_stolen'])

if 'gun_type' in df_gundeaths.columns:
    df_gundeaths = df_gundeaths.drop(columns=['gun_type'])

if 'latitude' in df_gundeaths.columns:
    df_gundeaths = df_gundeaths.drop(columns=['latitude'])

if 'location_description' in df_gundeaths.columns:
    df_gundeaths = df_gundeaths.drop(columns=['location_description'])

if 'notes' in df_gundeaths.columns:
    df_gundeaths = df_gundeaths.drop(columns=['notes'])

if 'n_guns_involved' in df_gundeaths.columns:
    df_gundeaths = df_gundeaths.drop(columns=['n_guns_involved'])

if 'sources' in df_gundeaths.columns:
    df_gundeaths = df_gundeaths.drop(columns=['sources'])

if 'incident_url_fields_missing' in df_gundeaths.columns:
    df_gundeaths = df_gundeaths.drop(columns=['incident_url_fields_missing'])


if 'incident_url' in df_gundeaths.columns:
    df_gundeaths = df_gundeaths.drop(columns=['incident_url'])

if 'source_url' in df_gundeaths.columns:
    df_gundeaths = df_gundeaths.drop(columns=['source_url'])
```
